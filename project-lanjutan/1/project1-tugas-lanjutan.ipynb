# Sistem Rekomendasi Video Youtube

```python
import pandas as pd
```

```python
# Source dataset https://www.kaggle.com/datasets/rsrishav/youtube-trending-video-dataset
file_path = "US_youtube_trending_data.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")
df.head()
```

```python
# cek data / Examining Data
print("Rows : ", df.shape[0])
print("Columns : ", df.shape[1])
print("Columns Features : ", df.columns.tolist())
print("Counf of Non Values : ", df.isnull().sum().values.sum())
```

```python
#  Check data with null value
df.isnull().sum()
```

```python
# Cek yang kosong pada kolom description
df.loc[df['description'].isnull()].head()
```

```python
# hapus data yg NA pada kolom description
df = df.dropna(subset='description')
print("Rows : ", df.shape[0])
```

```python
# hapus data yg mempunyai duplikasi pada video_id
df.drop_duplicates(subset='video_id', keep='last', inplace=True)
print("Rows : ", df.shape[0])
```

```python
# reset index supaya muncul index
df = df.reset_index()
```

```python
# setelah dilakukan reset maka muncul indexing kolom nya
df.head()
```

## Membuat Sistem Rekomendasi bedasarkan deskripsi

```python
from sklearn.feature_extraction.text import TfidfVectorizer
df = df[:2000]

#inisialisasi vectorizer
vectorize = TfidfVectorizer()

#ekstraksi fitur dari deskripsi video
fitur_deskripsi = vectorize.fit_transform(df['description'])
fitur_deskripsi
```

```python
# Untuk memeriksa kesamaan isi data
from sklearn.metrics.pairwise import cosine_similarity

description_in_matrix = pd.DataFrame(cosine_similarity(fitur_deskripsi))
description_in_matrix
```

```python
# 
description_in_matrix.columns = description_in_matrix.T.index
description_in_matrix['description'] = description_in_matrix.T.index
description_in_matrix = description_in_matrix.set_index('description')
description_in_matrix
```

```python
# coba disorting berdasarkan rekomendasi
top_10_recommended = list(description_in_matrix.loc[1995].sort_values(ascending=False).iloc[1:11].index)
top_10_recommended
```

```python
# top_10_recommended_product = df.loc[df.index.isin(top_10_recommended), ['title']].drop_duplicates().set_index('index').loc[top_10_recommended]
top_10_recommended_product = df.loc[df.index.isin(top_10_recommended), ['title', 'description']]
top_10_recommended_product
```

## Membuat Sistem Rekomendasi bedasarkan title

```python
from sklearn.feature_extraction.text import TfidfVectorizer
df = df[:2000]

#inisialisasi vectorizer
vectorize = TfidfVectorizer()

#ekstraksi fitur dari deskripsi video
fitur_title = vectorize.fit_transform(df['title'])
fitur_title
```

```python
# Untuk memeriksa kesamaan isi data
from sklearn.metrics.pairwise import cosine_similarity

tite_in_matrix = pd.DataFrame(cosine_similarity(fitur_title))
tite_in_matrix
```

```python
tite_in_matrix.columns = tite_in_matrix.index
tite_in_matrix['title'] = tite_in_matrix.index
tite_in_matrix = tite_in_matrix.set_index('title')

tite_in_matrix
```

```python
# coba disorting berdasarkan rekomendasi
tite_in_matrix.loc[2].sort_values(ascending=False)
```

```python
item_1995 = set(tite_in_matrix.loc[1995].iloc[tite_in_matrix.loc[1995].to_numpy().nonzero()].index)
```

```python
item_4 = set(tite_in_matrix.loc[4].iloc[tite_in_matrix.loc[4].to_numpy().nonzero()].index)
```

```python
top_10 = list(tite_in_matrix.loc[4].sort_values(ascending=False).iloc[1:11].index)
top_10
```

```python
top_10_recommended_videos = df.loc[df.index.isin(top_10), ['title', 'description']]
top_10_recommended_videos
```
