# Project 1 Sistem Rekomendasi Produk E-Commerce

## Import library & Load Data 

```python
import pandas as pd

# Source dataset https://www.kaggle.com/datasets/carrie1/ecommerce-data
dataFile = pd.read_csv("Project1-invoice-data.csv", encoding="ISO-8859-1")

dataFile
print('Rows: ', dataFile.shape[0])
print('Columns: ', dataFile.shape[1])
print('Columns Features: ', dataFile.columns.tolist())
print('Non Values: ', dataFile.isnull().sum().values.sum())
```

## Membersihkan Data

```python
dataFile.describe()
```

```python
dataFile = dataFile.loc[dataFile['Quantity'] > 0]
dataFile = dataFile.loc[dataFile['UnitPrice'] > 0]
dataFile.describe()
```

```python
dataFile.isnull().sum()
dataFile.loc[dataFile['CustomerID'].isna()].head()
```

```python
dataFile = dataFile.dropna(subset='CustomerID')
print('Rows:', dataFile.shape[0])
```

```python
customers_matrix = dataFile.pivot_table(index="CustomerID", columns="StockCode", values="Quantity", aggfunc="sum")
customers_matrix
```

```python
customers_matrix = (customers_matrix > 0).astype(int)
customers_matrix
```

## Mendapatkan Rekomendasi Produk berdasarkan Kode Produk

```python
from sklearn.metrics.pairwise import cosine_similarity

customers_in_matrix = pd.DataFrame(cosine_similarity(customers_matrix.T))
customers_in_matrix
```

```python
customers_in_matrix.columns = customers_matrix.T.index
customers_in_matrix['StockCode'] = customers_matrix.T.index
customers_in_matrix = customers_in_matrix.set_index('StockCode')

customers_in_matrix
```

```python
top_10_recommended = list(customers_in_matrix.loc["10124A"].sort_values(ascending=False).iloc[1:11].index)
top_10_recommended
```

```python
top_10_recommended_product = dataFile.loc[dataFile['StockCode'].isin(top_10_recommended), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode').loc[top_10_recommended]
top_10_recommended_product
```
